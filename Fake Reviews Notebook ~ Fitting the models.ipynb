## Fitting the FraudEagle

#### Creating the FraudEagle network graph

```python
# Create a graph with a parameter `epsilon`.
epsilon = 0.25
graph = feagle.ReviewGraph(epsilon)

# Extract relevant columns
review_data = concatenated_df[['User_ID', 'Product_ASIN', 'Review_Rating']]
```

```python
# Normalize the ratings
max_rating = review_data['Review_Rating'].max()
min_rating = review_data['Review_Rating'].min()

review_data['Normalized_Rating'] = (review_data['Review_Rating'] - min_rating) / (max_rating - min_rating)
```

```python
# Add reviewers and products to the graph
reviewers = {}
products = {}

for index, row in review_data.iterrows():
    reviewer_id = row['User_ID']
    product_id = row['Product_ASIN']
    rating = row['Normalized_Rating']
    
    # Add reviewer
    if reviewer_id not in reviewers:
        reviewers[reviewer_id] = graph.new_reviewer(reviewer_id)
    
    # Add product
    if product_id not in products:
        products[product_id] = graph.new_product(product_id)
    
    # Add review
    graph.add_review(reviewers[reviewer_id], products[product_id], rating)
```

```python
# Run analysis
print("Start iterations.")
max_iteration = 10000
for i in range(max_iteration):
    diff = graph.update()
    print(f"Iteration {i + 1} ends. (diff={diff})")
    if diff < 10**-5:
        break

# Retrieve results
reviewer_scores = {r.name: r.anomalous_score for r in graph.reviewers}
product_summaries = {p.name: p.summary for p in graph.products}
```

#### Displaying the FraudEagle review graphs

A simplified subset of the entire graph is displayed in a review graph

```python
# Simplified graph visualization
# Select a subset of the graph for visualization
subset_reviewers = list(reviewer_scores.keys())[:1000]
subset_products = list(product_summaries.keys())[:1000]

# Abbreviate IDs for visualization
abbreviate = lambda x: x[:6] + '...' if len(x) > 6 else x
abbreviated_reviewers = {reviewer: abbreviate(reviewer) for reviewer in subset_reviewers}
abbreviated_products = {product: abbreviate(product) for product in subset_products}

# Visualize the subset of the graph using networkx
G = nx.DiGraph()

# Add nodes
for reviewer in subset_reviewers:
    G.add_node(abbreviated_reviewers[reviewer], bipartite=0, color='blue')
for product in subset_products:
    G.add_node(abbreviated_products[product], bipartite=1, color='red')

# Add edges
for index, row in review_data.iterrows():
    if row['User_ID'] in subset_reviewers and row['Product_ASIN'] in subset_products:
        G.add_edge(abbreviated_reviewers[row['User_ID']], abbreviated_products[row['Product_ASIN']], weight=row['Normalized_Rating'])

# Draw the simplified graph
plt.figure(figsize=(20, 15))
pos = nx.spring_layout(G, k=0.1)  # k controls the distance between nodes
colors = ['blue' if data['bipartite'] == 0 else 'red' for _, data in G.nodes(data=True)]
weights = [G[u][v]['weight'] * 2 for u, v in G.edges]  # Smaller weight scaling for clarity

nx.draw(G, pos, with_labels=True, node_color=colors, edge_color='gray', width=weights, alpha=0.7, font_size=10)
plt.title("Simplified Review Graph Visualization")
plt.show()
```

## Fitting the SpEagle model

```python
# Step 1: Prepare the Data
graph_dict = {}
user_priors = {}
prod_priors = {}
review_priors = {}

# Convert the DataFrame to required formats
for index, row in concatenated_df.iterrows():
    user_id = row['User_ID']
    product_id = row['Product_ASIN']
    rating = row['Review_Rating']
    # Assign default label if not present
    label = row.get('label', -1)  # Use -1 as default if no label is available
    
    # Update user-product graph
    if user_id not in graph_dict:
        graph_dict[user_id] = {}
    if product_id not in graph_dict[user_id]:
        graph_dict[user_id][product_id] = {}
    graph_dict[user_id][product_id]['label'] = label

    # Update priors with example values
    if user_id not in user_priors:
        user_priors[user_id] = 0.5  # Example prior, adjust as needed
    if product_id not in prod_priors:
        prod_priors[product_id] = 0.5  # Example prior, adjust as needed

    review_id = (user_id, product_id)
    if review_id not in review_priors:
        review_priors[review_id] = 0.5  # Example prior, adjust as needed
```

```python
# Step 2: Define Edge Potentials
numerical_eps = 1e-5
user_review_potential = np.log(np.array([[1 - numerical_eps, numerical_eps], [numerical_eps, 1 - numerical_eps]]))
eps = 0.1
review_product_potential = np.log(np.array([[1 - eps, eps], [eps, 1 - eps]]))

potentials = {
    'u_r': user_review_potential, 
    'r_u': user_review_potential,
    'r_p': review_product_potential, 
    'p_r': review_product_potential
}

# Convert graph_dict to NetworkX graph
G = dict_to_networkx(graph_dict)


# Apply edge attribute filter to add labels to reviews
for u_id, products in graph_dict.items():
    for p_id, attributes in products.items():
        review_id = (u_id, p_id)
        G.add_edge(u_id, p_id, label=attributes['label'])

# Debugging: Check edges and nodes
print(f"Number of Nodes: {G.number_of_nodes()}")
print(f"Number of Edges: {G.number_of_edges()}")

# Check node types and priors
for node in G.nodes():
    if 'prior' not in G.nodes[node]:
        G.nodes[node]['prior'] = 0.5  # Ensure all nodes have a prior

# Check edge attributes
for u, v in G.edges():
    if 'label' not in G[u][v]:
        G[u][v]['label'] = -1  # Ensure all edges have a label
```

```python
# Step 3: Initialize and Run SpEagle
model = SpEagle(graph=G, potentials=potentials, max_iters=4)

# new runbp func
model.schedule(schedule_type='bfs')

# Debugging: Check the schedule
print(f"Schedule Length: {len(model._bp_schedule)}")
print(f"Belief Propagation Schedule: {model._bp_schedule}")

# Ensure the schedule is not empty
if len(model._bp_schedule) == 0:
    raise ValueError("Belief propagation schedule is empty. Check the graph and priors.")

# Run belief propagation with detailed debugging
try:
    model.run_bp(tol=1e-3)
except ZeroDivisionError:
    print("ZeroDivisionError encountered. Please check the schedule and total updates.")

# Classify nodes
user_scores, review_scores, product_scores = model.classify()

# Display results
print("User Scores:")
for user, score in user_scores.items():
    print(f"User {user}: {score}")

print("\nReview Scores:")
for review, score in review_scores.items():
    print(f"Review {review}: {score}")

print("\nProduct Scores:")
for product, score in product_scores.items():
    print(f"Product {product}: {score}")
```
